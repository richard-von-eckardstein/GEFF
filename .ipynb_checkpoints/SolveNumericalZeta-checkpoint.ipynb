```python
#Set up initial conditions for the GEF system (I plan to streamline this at some point)
alpha = 0.
Mpl = 1.
M = 6.16e-6*Mpl
phi0 = 15.55*Mpl
dphidt0 = -np.sqrt(2/3)*M*Mpl
ini = {"phi":phi0, "dphi":dphidt0}
ntr = 150
H = GEF(beta, Mpl, ini, M, ntr, approx=True)
```

```python
#Fix the kinetic coupling and its derivatives (normal kinetic term, no kinetic coupling)
I1 = 1
dI1 = 0
ddI1 = 0
```

```python
#Compute all auxiliary functions, e.g. the matrix M, the vector c_1 and c_2, the coefficients d^(\Phi) and
#   d^(\delta \phi) and finally p and q from the GEF results
M11 = 1

def M12(G):
    return -G.vals["H"]/G.vals["Hprime"]*(G.vals["dphi"]/2)/G.f #correct

def M21(G, k):
    a = G.vals["a"]
    
    prefac = G.omega*a*G.vals["H"]/G.vals["Hprime"]
    bracket = (k/a)**2 + G.ratio**2 * 2/3*I1*G.vals["B"]
    return bracket*prefac #correct 

def M22(G):
    a = G.vals["a"]
    E = G.vals["E"]
    B = G.vals["B"]
    dphi = G.vals["dphi"]
    H = G.vals["H"]
    
    prefac = G.ratio**3 * a * H/(G.vals["Hprime"]**2)
    t1 = dphi**2/6 * (dI1*E - G.dIdphi()*G.vals["G"])
    t2 = 1/3*I1*(E+B)*( H*dphi + G.dVdphi() + 1/6*dI1*G.ratio**2*(E + B) )
    return - prefac * (t1 + t2) #correct

def c1v(G):
    return -G.vals["H"]/(G.vals["Hprime"]*G.vals["a"]*G.omega)#correct
    
def c2rho(G):
    return 2/3*G.vals["H"]/(G.vals["Hprime"]*G.vals["a"]*G.omega) #correct
    
def c2pi(G, k):
    a = G.vals["a"]
    
    prefac = 2/3*G.omega*a*G.vals["H"]/G.vals["Hprime"]
    bracket = (k/a)**2 + G.ratio**2 * I1*G.vals["E"]
    return bracket*prefac #correct

def c2v(G):
    a = G.vals["a"]
    E = G.vals["E"]
    B = G.vals["B"]
    dphi = G.vals["dphi"]
    H = G.vals["H"]
    
    prefac =  H/G.vals["Hprime"]**2

    t1 = dphi * (G.dVdphi() - 1/3*G.ratio**2*(1/2*dI1*(E-B) - G.dIdphi()*G.vals["G"]))
    t2 = -2/3*I1*(E+B)*H*G.ratio**2
    return prefac * (t1 + t2) #correct

def Hddot(G):
    t1 = (G.vals["ddphi"] - G.dVdphi())*G.vals["dphi"]
    t2 = 1/6 * G.ratio**2 * (dI1*G.vals["dphi"]*(G.vals["E"]+G.vals["B"]) + I1*(G.vals["Edot"]+G.vals["Bdot"]) )
    return -(1/2)*(t1 + t2) - 3*G.vals["H"]*G.vals["Hprime"] #correct
    
def M21dot(G, k):
    d2Hdt2 = Hddot(G)
    dHdt = G.vals["Hprime"]
    H = G.vals["H"]
    a = G.vals["a"]

    t1 = (H**2/dHdt + (dHdt**2 - H*d2Hdt2)/dHdt**2)*((k/a)**2 + G.ratio**2*(2/3)*I1*G.vals["B"])
    t2 = H/dHdt * (-2*H*(k/a)**2 + G.ratio**2*(2/3) * (dI1*G.vals["dphi"]*G.vals["B"] + I1*G.vals["Bdot"]) )

    return a*G.omega*(t1 + t2)

def M22dot(G):
    d2Hdt2 = Hddot(G)
    dHdt = G.vals["Hprime"]
    H = G.vals["H"]
    a = G.vals["a"]
    
    E = G.vals["E"]
    B = G.vals["B"]
    EB = -G.vals["G"]
    Edot = G.vals["Edot"]
    Bdot = G.vals["Bdot"]
    EBdot = -G.vals["Gdot"]

    dphi = G.vals["dphi"]
    ddphi = G.vals["ddphi"]

    s1 = ( H*dphi + G.dVdphi() + 1/6*dI1*G.ratio**2*(E+B) )
    s2 = (dI1*E + G.dIdphi()*EB)

    t11 = dphi**2/6*s2
    t12 = 1/3*I1*(E + B)*s1
    t1 = ( (H/dHdt)**2 + (dHdt**2 - 2*H*d2Hdt2)/dHdt**3)*(t11 + t12)

    t21 = dphi*ddphi/3*s2
    t22 = dphi**2/6*(ddI1*dphi*E + dI1*Edot + G.ddIddphi()*dphi*EB + G.dIdphi()*EBdot)
    t23 = 1/3*(dI1*dphi*(E+B) + I1*(Edot + Bdot) )*s1
    t24 = 1/3*I1*(E+B)*(H*ddphi + dphi*(dHdt + G.ddVddphi() + 1/6*ddI1*G.ratio**2*(E+B))
                        +1/6*dI1*G.ratio**2*(Edot + Bdot))
    t2 = H/dHdt**2*(t21 + t22 + t23 +t24)

    return -G.ratio**3*a*(t1 + t2) #correct

def dPhi(G,k):
    H = G.vals["H"]
    a = G.vals["a"]
    
    E = G.vals["E"]
    B = G.vals["B"]
    EB = -G.vals["G"]
    
    dphi = G.vals["dphi"]

    t1 = M21dot(G,k) - H*M21(G,k) + 2*G.f/dphi*( -(k/a)**2 + dphi**2/2 - G.ratio**2*I1*B)*M22(G)
    t2 = a**2*G.ratio**2*G.omega**2*I1*(E+2*B)*c2v(G)/3

    t13 = (dI1*E + G.dIdphi()*EB)*( (k/a)**2 + G.ratio**2*I1*B)
    t23 = -dphi**2/2*(dI1*E + 3*G.dIdphi()*EB)
    t33 = -I1*dphi*(H*E + 2*G.vals["G1"])
    t3 = a**3*G.ratio**2*G.omega**3 * (t13 + t23 + t33)*c2rho(G)/dphi

    return G.omega*a*t1 + t2 + t3 #correct!

def ddelphi(G, k):
    H = G.vals["H"]
    a = G.vals["a"]
    
    E = G.vals["E"]
    B = G.vals["B"]
    EB = - G.vals["G"]
    
    dphi = G.vals["dphi"]

    t1 = M22dot(G) + dphi/(2*G.f)*M21(G,k) - (3*H + 1/dphi*(G.dVdphi() + 1/2*G.ratio**2*dI1*(E+B) ))*M22(G)
    t2 = -a/6*G.ratio**3*(dI1*(E - 2*B) + 3*G.dIdphi()*EB)*c2v(G)

    t13 = (dI1*E + G.dIdphi()*EB)*(3*H + 1/dphi*( G.dVdphi() + 1/2*G.ratio**2*dI1*(E+B) ) )
    t23 = -dphi*((ddI1 - dI1**2/I1)*E + (G.ddIddphi() - dI1*G.dIdphi()/I1)*EB)
    t33 = dphi*I1*E
    t3 = G.ratio**3*a**2*(t13 + t23 + t33)/2*c2rho(G)*G.omega

    return G.omega*a*(t1 + t2 + t3)

def pcoeff(G, k):
    detM = M11*M22(G) - M12(G)*M21(G,k)
    Minv12 = -M12(G)/detM
    Minv22 = M11/detM
    return (-dPhi(G,k)*Minv12 - ddelphi(G,k)*Minv22)/G.omega

def qcoeff(G,k):
    detM = M11*M22(G) - M12(G)*M21(G,k)
    Minv11 = M22(G)/detM
    Minv21 = -M21(G,k)/detM
    q = (-dPhi(G,k)*Minv11 - ddelphi(G,k)*Minv21)/G.omega**2
    return q
    
```

```python
def ZetaEoM(Z, k, p, q, a):
    #The equation of motion we want to solve
    dZdt = np.zeros(Z.shape)
    
    #real
    dZdt[0] = k/a*(Z[1] - Z[2])
    dZdt[1] = (- p*Z[1] - k*Z[3] - q/k*Z[0])/a
    
    #imaginary
    dZdt[2] = k/a*(Z[3] + Z[0])
    dZdt[3] = (- p*Z[3] + k*Z[1] - q/k*Z[2])/a
    
    return dZdt


def SolveZeta(G, tmax, steps=200, cut=0.1, NoBR=False):
    #For a given time span from t0 to tmax and for a fixed number of modes = steps:
    #   compute the solution for every wavenumber k=10^(5/2) a(t)H(t)
    #   where t in [t0, tmax]. 10^(5/2) guarentees that the mode is still deep inside the Horizon and
    #   can be initialised as Bunch--Davies
    N = G.vals["N"]
    t = G.vals["t"]
    G.Unitless()
    Nend = G.EndOfInflation()[0]
    tend = min(CubicSpline(N, t)(Nend), max(t))
    lend = 0
    while t[lend]<tend:
        lend+=1
    a = CubicSpline(t, G.vals["a"])
    kH = CubicSpline(t, G.vals["H"]*G.vals["a"])
    if NoBR:
        # for comparing with the no BR case
        H.vals["E"] = 0
        H.vals["B"] = 0
        H.vals["G"] = 0
        H.vals["G1"] = 0
        H.vals["Edot"] = 0
        H.vals["Bdot"] = 0
        H.vals["Gdot"] = 0
        H.vals["ddphi"] = 0
    z = G.vals["a"]*G.vals["dphi"]/G.vals["H"]
    zf = CubicSpline(t, G.vals["a"]*G.vals["dphi"]/G.vals["H"])
    zprime = G.vals["a"]*(G.vals["dphi"] + G.vals["ddphi"]/G.vals["H"] - 
               G.vals["dphi"]/G.vals["H"]**2*G.vals["Hprime"])
    damp = CubicSpline(t,H.vals["a"]*zprime/z)
    teval = np.linspace(2, tmax, steps)

    Zs = []
    Zprimes = []
    ks = []
    ts = []
    tstart = []
    l = 0
    for i, s in enumerate(teval):
        #wavenumber in question for a given mode
        k = 10**(5/2)*kH(s)
        ks.append(k) 
        while (t[l] < s):
            l+=1
        tstart.append(s)
        #p and q for given k
        p = CubicSpline(t, pcoeff(G, k))
        q = CubicSpline(t, qcoeff(G, k))
        
        #Initial conditions
        Zini = np.array([1, -damp(s)/k, 0, -1])

        ode = lambda x, y: ZetaEoM(y, k, p(x), q(x), a(x))

        sol = solve_ivp(ode, [s, tend], Zini, t_eval=t[l:lend], method="RK45", atol=1e-8, rtol=1e-8)
        print(sol.success)
        tst  = sol.t[0]
        
        #ensure that Zs is the same length for every mode k
        Zs.append( np.array([a(s)/(zf(s)*a(t[j])) for j in range(l)]
                  + list((sol.y[0,:] + 1j*sol.y[2,:])/zf(s)) ))
        
        
        
    return t[:lend], ks, Zs, tstart, tend
```
