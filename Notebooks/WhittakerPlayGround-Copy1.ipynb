```python
Ef = lambda x, xi, l: np.exp(xi*np.pi*l)*abs(1j*(2*abs(xi)*x - xi*l)*whitw(-1j*xi*l, 1/2, -4j*abs(xi)*x)
                                               + whitw(1-1j*xi*l, 1/2, -4j*abs(xi)*x) )**2/(2*abs(xi)*x)**2
Bf = lambda x, xi, l: np.exp(xi*np.pi*l)*abs( whitw(-1j*xi*l, 1/2, -4j*abs(xi)*x) )**2
Gf = lambda x, xi, l: np.exp(xi*np.pi*l)*l*(whitw(1-1j*xi*l, 1/2, -4j*abs(xi)*x)
                                            *whitw(-1j*xi*l, 1/2, -4j*abs(xi)*x).conjugate()).real/(2*abs(xi)*x)
```

```python
iE = lambda x, xi, l, n: x**(3+n)*Ef(x, xi, l)
iB = lambda x, xi, l, n: x**(3+n)*Bf(x, xi, l)
iG = lambda x, xi, l, n: x**(3+n)*Gf(x, xi, l)

def Exact(x, xi, l, n, name):
    if name=="E": val = x**3*Ef(x, xi, l)
    elif name=="B": val = x**3*Bf(x, xi, l)
    elif name=="G": val = x**3*Gf(x, xi, l)
    else: print("Error")
    return val

def Taylor(xi, pol, order):
    Y = {"E":np.zeros(order), "B":np.zeros(order), "G":np.zeros(order)}

    E = Ef(1, xi, pol)
    G = Gf(1, xi, pol)
    B = Bf(1, xi, pol)
    
    Y["E"][0] = E
    Y["B"][0] = B
    Y["G"][0] = G

    r = -2*abs(xi)
    p = (1 + 2*pol*xi/r)

    if order > 1:
        Y["E"][1] = -2*p*G*pol*r
        Y["B"][1] = 2*G*pol*r
        Y["G"][1] = pol*(E - p*B)*r
        sum1 = 0
        sum2 = 0
        for n in range(2,order):
            sum1 += (-1)**(n)*(n-1)*Y["B"][n-1]
            sum2 += (-1)**n*Y["B"][n-2]
            Bn = 2*pol*r/n*Y["G"][n-1]
            En = -p*Bn + 2*pol*xi/(r*n)*(-1)**n*sum1
            Gn = pol*r/n*(Y["E"][n-1] - p*Y["B"][n-1]) + (-1)**n*2*xi/n*sum2
            Y["E"][n] = En
            Y["B"][n] = Bn
            Y["G"][n] = Gn

    return Y

def IntApprx(xs, xi, l, order=3, n=0, mode=None):
    ys = xs-1
    
    taylor = Taylor(xi, l, order)

    coeff = np.zeros(order)
    for i in range(min(n+4, order)):
        coeff[i] = binom(n+3, i)
    #print(coeff)
        
    dic = {"E":0, "B":0, "G":0}
    for name in dic.keys():
        for i in range(order):
            for j in range(i+1):
                dic[name] += (-1)*(ys)**(i+1)/(i+1) * (taylor[name][j]*coeff[i-j])
    return dic


def IntegrandApprx(xs, xi, l, order=3, n=0, mode=None):
    ys = xs-1
    
    taylor = Taylor(xi, l, order)
    
    coeff = np.zeros(order)
    for i in range(min(n+4, order)):
        coeff[i] = binom(n+3, i)
    #print(coeff)
        
    dic = {"E":0, "B":0, "G":0}
    for name in dic.keys():
        for i in range(order):
            for j in range(i+1):
                dic[name] += taylor[name][j]*coeff[i-j] * (ys)**(i)
    return dic
```
